```python
from transformers import pipeline

# Initialize the question-answering pipeline using a pre-trained model
qa_pipeline = pipeline('question-answering', model='distilbert-base-cased-distilled-squad', tokenizer='distilbert-base-cased')
```

```python
#FOR TEXT BASED QUESTION 
global options

def ask_question(topic):
    global options

    topic_questions = df[df['Topic'] == topic]
    
    if topic_questions.empty:
        return "No questions available for the selected topic."
    
    random_question = topic_questions.sample(n=1)
    question = random_question.iloc[0]['Question']
    answer = random_question.iloc[0]['Answer']
    #print(answer)
    options = random_question.iloc[0]['Options']
    
    print(f"Question: {question}")
    
    # Display multiple-choice options
    print("Options:",options)
    
    options_list = options.split()
    correct_answer_index = options_list.index(answer)
    correct_option = options_list[correct_answer_index - 1][0]
    print(correct_option)  # Output

    
    user_answer = input("Your Answer (enter the letter corresponding to the correct option): ")
    
    # Map user's answer (letter) to the corresponding option
    #answer_mapping = {'a': 'A', 'b': 'B', 'c': 'C', 'd': 'D'}
    user_option = user_answer.lower()
    
    # Use the model to predict the answer
    model_prediction = qa_pipeline(context=question, question=user_option)
    predicted_answer = model_prediction['answer']

    print("-------------------------------------------------------------------------------------------------")
    if user_option.lower() == correct_option.lower():
        print("WOAH! Your answer is correct -> ", answer) 
        #print("WOAH! Your answer is correct -> ", correct_option) 
        
    else:
        return f"Incorrect! The correct answer is: {correct_option}. {answer}"

def main():
    
    available_topics = df['Topic'].unique()
    print("Welcome to the Question-Answering System!")
    print("Available Topics: ",available_topics)  # Update available topics as needed
    
    # Prompt user to choose a topic
    selected_topic = input("Select a topic from the available options: ")  # Update available topics as needed
    
    # Validate the selected topic  # Update available topics as needed
    if selected_topic not in available_topics:
        print("Invalid topic selection. Please choose from the available options.")
        return
    
    result = ask_question(selected_topic)
    
main()
```
